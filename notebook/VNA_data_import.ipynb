## Testing import with read_s4p

```python
from io_utils.s2p import read_s2p
from io_utils.s4p import read_s4p

file_path = "../data/2021_03_18_-_TD31_Nx-raw_data/01-Before_Tuning/01-SP-N2-wire.s4p"
[f, s11, s12, s21, s22] = read_s4p(file_path)
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
ax.plot(f, np.abs(s11), label="s11")
ax.set_ylabel("Magnitude [Units]")
ax.set_xlabel("Frequency [Hz]")

ax.grid(True, linestyle="--", alpha=0.5)
ax.legend()
```
